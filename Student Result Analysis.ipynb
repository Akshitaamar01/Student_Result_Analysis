```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Students_scores.csv")
print(df.head())
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

## Drop unnamed column

```python
df = df.drop("Unnamed: 0", axis=1)
print(df.head())
```

 ## Change weekly study hours columns

```python
df["WklyStudyHours"] = df["WklyStudyHours"].str.replace("05-Oct","5-10")
df.head()
```

## Gender Distribution

```python
plt.figure(figsize=(5,5))
ax = sns.countplot(data=df, x="Gender")
ax.bar_label(ax.containers[0])
plt.title('Gender Distribution')
plt.show()
```

### From the above chart, we have analyzed that the number of females in the data is more than the number of males

```python
gb = df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print (gb)
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb, annot = True)
plt.title("Relationship between Parent's Education and Student's Score")
plt.show()
```

### From the above chat we have concluded that the education of the parents have a good impact on the scores of the student

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print (gb1)
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb1, annot = True)
plt.title("Relationship betweem Parent's Marital Status and Student's score")
plt.show()
```

### From the above chart we have analyzed that the marital status of the parent has negligible impact on the scores of the students
